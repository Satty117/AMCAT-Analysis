GENDER - SPECIALIZATION RELATION

```python
from scipy.stats import chi2_contingency


contingency_table = pd.crosstab(data['Gender'], data['Specialization'])


chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square test p-value: {p_value}")

if p_value < 0.05:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("No significant relationship between Gender and Specialization.")
```
